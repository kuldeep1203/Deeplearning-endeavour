```python
import numpy as  np 
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline 
from tensorflow import keras 
```

```python
dataframe = pd.read_csv("./datasets/insurance_data(1).csv")
```

```python
dataframe
```

```python
#we could standardize the data  of age btwn 1 and 0 ideally the age shld be between 0 and 100
from sklearn.model_selection import train_test_split
X_train , X_test ,y_train ,y_test = train_test_split(dataframe[['age','affordibility']],dataframe.bought_insurance,test_size=0.2,random_state=25)
```

```python
X_train_scaled = X_train.copy()
X_train_scaled['age'] = X_train_scaled['age']/100
X_test_scaled = X_test.copy()
X_test_scaled['age'] = X_test_scaled['age']/100
```

```python
X_train_scaled
```

```python
#two input parameters 
#intializing weights to be 1 and bias to be  0
model = keras.Sequential([
    keras.layers.Dense(1,input_shape=(2,),activation="sigmoid",kernel_initializer='ones',bias_initializer='zeros')
])
```

```python
model.compile(
    optimizer ='adam',
    loss = 'binary_crossentropy',
    metrics = ['accuracy']
)
```

```python
model.fit(X_train_scaled,y_train,epochs=5000)
```

```python
model.evaluate(X_test_scaled,y_test)
```

```python
coef,intercept = model.get_weights()
coef,intercept
```

```python
def sigmoid(x):
    import math
    return 1/(1+math.exp(-x))
sigmoid(18)
```

```python
def prediction_function(age,affordibility):
    weighted_sum =  coef[0]*age+coef[1]*affordibility + intercept # intercept is bias
    return sigmoid(weighted_sum)
```

```python
def log_loss(y_true,y_predicted):
    epsilon=1e-15
    y_predicted_new = [max(i,epsilon) for i in y_predicted]
    y_predicted_new = [min(i,1-epsilon) for i in y_predicted_new]
    y_predicted_new = np.array(y_predicted_new)
    return -np.mean(y_true*np.log(y_predicted_new)+(1-y_true)*np.log(1-y_predicted_new))
```

```python
def sigmoid_numpy(X):
    return 1/(+np.exp(-X))
sigmoid_numpy(np.array([12,0,1]))
```

```python
def gradient_descent(age,affordability,y_true,epochs):
    #w1 w2 bias
    w1 = w2 = 1
    bias = 0
    rate = 0.5
    n = len(age)

    for i in  range (epochs):
        #these are applied on numpy vectors
        weighted_sum  = w1*age + w2*affordability + bias
        y_predicted = sigmoid_numpy(weighted_sum)
        loss = log_loss(y_true,y_predicted)

        w1d = (1/n)*np.dot(np.transpose(age),(y_predicted-y_true))
        w2d = (1/n)*np.dot(np.transpose(affordability),(y_predicted-y_true)) 

        biasd = np.mean(y_predicted-y_true)

        w1 = w1 - rate*w1d
        w2 = w2-rate*w2d
        bias = bias - rate*biasd

        print(f"Epoch:{i},w2:{w2},bias:{bias},loss:{loss}")

    return w1,w2,bias
        
```
